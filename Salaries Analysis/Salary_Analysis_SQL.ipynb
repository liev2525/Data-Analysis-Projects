## DATA ANALYSIS

```python
%sql
CREATE TABLE salaries_raw (
    work_year INT,
    experience_level VARCHAR(2),
    employment_type VARCHAR(2),
    job_title STRING,
    salary INT,
    salary_currency VARCHAR(5),
    salary_in_usd INT,
    employee_residence VARCHAR(2),
    remote_ratio INT,
    company_location VARCHAR(2),
    company_size VARCHAR(1)
);
```

```python
%sql
COPY INTO salaries_raw
FROM '/Volumes/dp_analysis/dp_analysis_1/dp_vol1/salaries.csv'
FILEFORMAT = CSV
FORMAT_OPTIONS (
  'header' = 'true',
  'inferSchema' = 'true'
);
```

```python
%sql
CREATE OR REPLACE TABLE salaries_clean AS
SELECT
    work_year,
    CASE experience_level
        WHEN 'EN' THEN 'Entry'
        WHEN 'MI' THEN 'Mid'
        WHEN 'SE' THEN 'Senior'
        WHEN 'EX' THEN 'Executive'
    END AS experience_level,
    CASE employment_type
        WHEN 'FT' THEN 'Full-Time'
        WHEN 'PT' THEN 'Part-Time'
        WHEN 'CT' THEN 'Contract'
    END AS employment_type,
    job_title,
    salary_in_usd,
    employee_residence,
    remote_ratio,
    company_location,
    CASE company_size
        WHEN 'S' THEN 'Small'
        WHEN 'M' THEN 'Medium'
        WHEN 'L' THEN 'Large'
    END AS company_size
FROM salaries_raw
WHERE salary_in_usd IS NOT NULL;

SELECT * FROM salaries_clean;
```

```python
%sql
CREATE OR REPLACE TABLE salary_gold_experience AS
SELECT
    experience_level,
    ROUND(AVG(salary_in_usd), 2) AS avg_salary_usd,
    COUNT(*) AS employee_count
FROM salaries_clean
GROUP BY experience_level;

SELECT * FROM salary_gold_experience
ORDER BY avg_salary_usd DESC
LIMIT 10;
```

```python
%sql
CREATE OR REPLACE TABLE salary_gold_job_title AS
SELECT
    job_title,
    ROUND(AVG(salary_in_usd), 2) AS avg_salary_usd,
    COUNT(*) AS role_count
FROM salaries_clean
GROUP BY job_title
HAVING COUNT(*) >= 10;

SELECT * FROM salary_gold_job_title
ORDER BY avg_salary_usd DESC
LIMIT 10;
```

```python
%sql
CREATE OR REPLACE TABLE salary_gold_remote AS
SELECT
    CASE remote_ratio
        WHEN 0 THEN 'On-Site'
        WHEN 50 THEN 'Hybrid'
        WHEN 100 THEN 'Fully Remote'
    END AS work_type,
    ROUND(AVG(salary_in_usd), 2) AS avg_salary_usd,
    COUNT(*) AS employee_count
FROM salaries_clean
GROUP BY remote_ratio;

select * from salary_gold_remote
LIMIT 10;
```

```python
%sql
CREATE OR REPLACE TABLE salary_gold_country AS SELECT
    CASE company_location
        WHEN 'US' THEN 'United States'
        WHEN 'CA' THEN 'Canada'
        WHEN 'GB' THEN 'United Kingdom'
        WHEN 'DE' THEN 'Germany'
        WHEN 'FR' THEN 'France'
        WHEN 'IN' THEN 'India'
        WHEN 'ES' THEN 'Spain'
        WHEN 'AU' THEN 'Australia'
        WHEN 'NL' THEN 'Netherlands'
        WHEN 'IT' THEN 'Italy'
        WHEN 'BR' THEN 'Brazil'
        WHEN 'PL' THEN 'Poland'
        WHEN 'PT' THEN 'Portugal'
        WHEN 'RU' THEN 'Russia'
        WHEN 'IE' THEN 'Ireland'
        WHEN 'CH' THEN 'Switzerland'
        WHEN 'UA' THEN 'Ukraine'
        WHEN 'AT' THEN 'Austria'
        WHEN 'BE' THEN 'Belgium'
        WHEN 'MX' THEN 'Mexico'
        WHEN 'RO' THEN 'Romania'
        WHEN 'GR' THEN 'Greece'
        WHEN 'TR' THEN 'Turkey'
        WHEN 'SG' THEN 'Singapore'
        WHEN 'AR' THEN 'Argentina'
        WHEN 'CL' THEN 'Chile'
        WHEN 'DK' THEN 'Denmark'
        WHEN 'FI' THEN 'Finland'
        WHEN 'SE' THEN 'Sweden'
        WHEN 'CZ' THEN 'Czech Republic'
        WHEN 'HU' THEN 'Hungary'
        WHEN 'IL' THEN 'Israel'
        WHEN 'ZA' THEN 'South Africa'
        WHEN 'NO' THEN 'Norway'
        WHEN 'EE' THEN 'Estonia'
        WHEN 'LT' THEN 'Lithuania'
        WHEN 'SK' THEN 'Slovakia'
        WHEN 'BG' THEN 'Bulgaria'
        WHEN 'HR' THEN 'Croatia'
        WHEN 'SI' THEN 'Slovenia'
        WHEN 'LU' THEN 'Luxembourg'
        WHEN 'LV' THEN 'Latvia'
        WHEN 'NZ' THEN 'New Zealand'
        WHEN 'JP' THEN 'Japan'
        WHEN 'CN' THEN 'China'
        WHEN 'AE' THEN 'United Arab Emirates'
        WHEN 'TH' THEN 'Thailand'
        WHEN 'MY' THEN 'Malaysia'
        WHEN 'KR' THEN 'South Korea'
        WHEN 'EG' THEN 'Egypt'
        WHEN 'NG' THEN 'Nigeria'
        WHEN 'CO' THEN 'Colombia'
        WHEN 'PE' THEN 'Peru'
        WHEN 'PH' THEN 'Philippines'
        WHEN 'ID' THEN 'Indonesia'
        WHEN 'SA' THEN 'Saudi Arabia'
        WHEN 'PK' THEN 'Pakistan'
        ELSE company_location
    END AS country,
    ROUND(AVG(salary_in_usd), 2) AS avg_salary_usd,
    COUNT(*) AS role_count
FROM salaries_clean
GROUP BY company_location
HAVING COUNT(*) >= 20;

SELECT * FROM salary_gold_country
ORDER BY avg_salary_usd DESC
LIMIT 10;
```
